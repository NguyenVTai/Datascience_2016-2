```python
from requests_html import HTMLSession, HTML
from selenium import webdriver
import time
import re
```

```python
session = HTMLSession()
```

```python
browser = webdriver.Chrome(executable_path='chromedriver.exe')
```

```python
#danh sach cac nam muon lay du lieu
#nhom chi lay 1 nam la 2018 truoc de kiem tra mo hinh
list_year = [2018]
```

```python
filename='wth_data.csv'
```

```python
list_w = set()
with open(filename,'w') as f:
        f.write('time\tmonth\ttemperature\twind\thumidity\tbarometer\tvisibility\tweather\n')
```

```python
def crawl_in_page (list_year,filename):
    for year in list_year:
        for month in range(12):
            month+=1
            url=f'https://www.timeanddate.com/weather/vietnam/ho-chi-minh/historic?month={month}&year={year}'
            browser.get(url)
            
            days = browser.find_elements_by_xpath("//div[@class='row pdflexi']/div[@class='weatherLinks']/a")
            for i in range(len(days)):
                days[i].click()
                time.sleep(2)
                r = HTML(html=browser.page_source)
                print('data ngay: ', r.find('.mgt0')[1].text )
                table = r.find('table')[1]
                crawl_in_table(table,month,filename)
```

```python
def crawl_in_table (table,month,filename):
    data = table.find('tbody',first = True)
    trs = data.find('tr')

    for tr in trs:
        #rut trich gio
        time = tr.find('th',first=True).text
        time = re.findall(r'\b\d{1,2}:\d{1,2}\b',time)[0]
        h1 = float(re.findall(r'\b(\d{1,2}):\w+',time)[0])
        h2 = re.findall(r'\b:(\d{1,2})',time)[0]
        if(h2 == '30'):
            h1+=0.5
        #thang
        #month
        
        
        tds=tr.find('td')
        
        #nhiet do
        temp = int(tds[1].text.replace('\xa0°C',''))
    
        #suc gio
        
        wind = tds[3].text
        if ((wind == 'No wind') | (wind =='N/A')):
            wind = 0
        else:
            wind = int(wind.replace(' km/h',''))
        #do am
        humidity= tds[5].text
        if (humidity =='N/A'):
            humidity=0
        else:
            humidity= int(humidity.replace('%',''))
    
        #ap suat
        barometer = tds[6].text
        if (barometer=='N/A'):
            barometer=0
        else:
            barometer= int(barometer.replace(' mbar','')) 
        #tam nhin xa
        visibility = (tds[7].text)
        if (visibility == 'N/A'):
            visibility = 0
        else:
            visibility = int(visibility.replace('\xa0km',''))
        #weather
        weather =tds[2].text.replace('.','')
        list_w.add(weather)
        #ghi ra file csv
        with open(filename,'a') as f:
            f.write(f'{h1}\t{month}\t{temp}\t{wind}\t{humidity}\t{barometer}\t{visibility}\t{weather}\n')
```

```python
crawl_in_page(list_year,filename)
```
